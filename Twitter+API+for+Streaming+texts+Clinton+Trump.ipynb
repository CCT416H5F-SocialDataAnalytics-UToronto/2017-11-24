```python
##We need to face the possibilities of errors, and this line of codes help in that way


class StreamListener(tweepy.StreamListener):

##Initialiser of the class
##number of tweets
##number of characters
##file object
##polarities scores for Donald Trump and Hillary Clinton
##initializing of the parent class

    def __init__(self):
        self.num_tweets = 0
        self.num_characters = 0
        self.file = open('tweets.txt', 'a')
        self.polarities = {
            'Donald Trump': [],
            'Hillary Clinton': []
        }
        super().__init__()

##definition of the status
##writing the tweet in the file
##we close the file
##we increment the number of tweets and characters

    def on_status(self, status):
        
        self.file = open('tweets.txt', 'a')
        self.file.write('{}'.format(status.text))
        self.file.close()
        self.num_tweets += 1
        self.num_characters += len(status.text)

##We analyze the text using the textblob
##storing the polarities scores for Donald Trump and Hillary Clinton
##we print the sentiment scores and the tweets texts
        b = TextBlob(status.text)
        if 'trump' in status.text.lower():
            self.polarities['Donald Trump'].append(b.sentiment.polarity)
        elif 'clinton' in status.text.lower():
            self.polarities['Hillary Clinton'].append(b.sentiment.polarity)
        print(b.sentiment)
        print("Tweet {}: {}".format(self.num_tweets, status.text))


##Calling this function to catch the error
    def on_error(self, status_code):
        if status_code == 420:
            return False

##We now start the listener
stream_listener = StreamListener()
stream = tweepy.Stream(auth=api.auth, listener=stream_listener)
stream.filter(track=["trump", "clinton", "hillary clinton", "donald trump"])
```
